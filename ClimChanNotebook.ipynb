# The effects of climate change on global crop yields

Climate change will play a major role in global crop production.  As climate change causes extreme weather conditions such as drought, some areas of the world which are prone to famine will see their crop yields decrease dramatically.  The following is an analysis of the drop in crop yeilds using different warming scenarios based on CO2 emissions.  The dataset holds information of the projected drop-off in production for three different crops:  Rice, Wheat and Maize.  The following plots pertain to Rice production in 165 countries in order to give a global perspective of the potential impact of climate change.  Also, it is worth noting that the graphs utilize the worst-case climate change scenario in which the CO2 levels are projected to be the highest.  The first plot is the baseline and shows the global average crop production per year from 2000 to 2006.  The second plot shows the change in crop production from the baseline for the worst-case climate change scenario projected to 2050.  The last plot shows the change in crop production from the baseline for the worst-case climate change scenario projected to 2080.



```r

library(readxl)
library(dplyr)
library(ggplot2)
options(digits=10)
options(warn = -1)

data <- read_xlsx("C:\\Users\\James\\source\\repos\\RProjects\\CapstoneData\\Copy_of_crops-yield-changes-hadcm3-sresDataOnly.xlsx")





#data <- na.omit(data)
riceData <- data.frame(stringsAsFactors = FALSE, data$Fips_code, data$ISO3v10, data$RI_2000, data$RIA1F2020, data$ActChRIAIF2050, data$ActChRIAIF2080, data$ActChRIA2a2020, data$ActChRIA2a2050, data$ActChRIA2a2080, data$ActChRIA2b2020, 
                       data$ActChRIA2b2050, data$ActChRIA2b2080, data$ActChRIA2c2020, data$ActChRIA2c2050, data$ActChRIA2c2080, data$ActChRIB1a2020, data$ActChRIB1a2050, data$ActChRIB1a2080, 
                       data$ActChRIB2a2020, data$ActChRIB2a2050, data$ActChRIB2a2080, data$ActChRIB2b2020, data$ActChRIB2b2050, data$ActChRIB2b2080)

newData <- data.frame(riceData$data.Fips_code, riceData$data.RI_2000)
secondData <- data.frame(riceData$data.Fips_code, riceData$data.ActChRIAIF2050)
thirdData <- data.frame(riceData$data.Fips_code, riceData$data.ActChRIAIF2080)
delta1 <- c()
delta2 <- c()
for(i in 1:nrow(newData)){
  delta1[length(delta1) + 1] <- newData[i,2]  + secondData[i,2]
}

for(i in 1:nrow(newData)){
  delta2[length(delta2) + 1] <- newData[i,2]  + thirdData[i,2]
}
#print(delta1)
#print(delta2)
x = newData$riceData.data.Fips_code

y = newData$riceData.data.RI_2000
secondy = delta1
thirdy = delta2
plot(y , names.arg = x, type="l", ylab="Average Rice production in tons", xlab="Countries", main="2000 To 2006" )
plot(secondy , names.arg = x, type="l", ylab="Projected Rice production in tons", xlab="Countries", main="2050 Projection")
plot(thirdy , names.arg = x, type="l", ylab="Projected Rice production in tons", xlab="Countries", main="2080 Projection")

```
