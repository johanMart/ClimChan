# Land use from world crop production and the case for vertical farming.

Population projections predict that by 2050, there will be close to 10 billion people on the planet.  This projection brings many questions to the forefront.  Is our way of life sustainable when we have close to 10 billion people on Earth?  Will our current farming techniques suffice to feed the globe?  How will global warming force us to change the way we live in order to save the planet?

When questions like these are asked, we have to take a good hard look at the data in order to make decisions that will impact our ability to survive on this planet in the future.  It turns out that our current way of going about feeding ourselves will not be effective in the coming decades.  As of right now, about 50% of all habitable land (land that is not covered by glaciers, or deserts, dry salt flats, beaches, sand dunes, and exposed rocks) is used for agriculture.  Of this 50%, about 25% or 15,925,836(in 2016 when the world population was 7,418,151,841)  square km is used to grow crops(not livestock).  As the world population grows(about 26% growth from now to 2050) we need to come up with more and more efficient uses of our land for subsistence.  

One technology which is believed will be able to allow us to use our resources more efficiently is vertical farming. Vertical farming refers to the idea that crops can be grown in a high rise, under optimal conditions, with very close observation.  This idea, if realized correctly, can solve both the land use problem and the freshwater usage problem we have with traditional farms of today.  The gist is that a vertical farm would have a smaller square footage land footprint than a comparable farm in terms of production because it would stack its space vertically.  

Besides alleviating the strain on the planet for land use for growing crops, vertical farming also helps preserve freshwater, as using hydroponic systems that recycle water have resulted in some vertical farms touting 95% less freshwater usage than traditional farms. Vertical farms might also lower greenhouse gas emissions produced by farming because if the vertical farm is built in an urban area, transporting the crops to the city will reduce the distance the crops are on the road.  There are a myriad of factors why vertical farms are a viable alternative to traditional farming, and there are also many hurdles, and upfront costs, financially and enivronmentally, that come attached to the production of a vertical farm.

For this project in particular, I will focus on how vertical farms can help the planet from the perspective of land use and freshwater withdrawal.  The project will aim to use population projections to predict global food needs and in turn land usage and freshwater usage requirements for the food needs.  The assignment will also aim to highlight areas with the most rapid growing populations in order to advise where the vertical farms will be useful.
